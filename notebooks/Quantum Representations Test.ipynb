#### Quantum Representations



##### Example Data

```python
original_data = utils.simulate_data(num_samples=8)
utils.plot(original_data)
```

##### A) Quantum Probability Amplitude Modulation (QPAM)

```python
scheme_a = qpam.QPAM()
```

```python
encoded_circuit_a = scheme_a.encode(original_data)
```

```python
encoded_circuit_a.draw('mpl',style='clifford')
```

```python
decoded_data = scheme_a.decode(encoded_circuit_a)
```

```python
utils.plot([original_data,decoded_data],title='QPAM')
```

```python
# interactive reconstruction of signal
#utils.tune(encoded_circuit_a,function=scheme_a.decode,step=2,max_value=4000,ref=original_data)
```
