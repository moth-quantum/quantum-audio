#### Quantum Representations



##### Example Data

```python
original_data = utils.simulate_data(num_samples=8)
label = ('original data','reconstructed data')
utils.plot(original_data,label=label[0])
```

##### A) Quantum Probability Amplitude Modulation (QPAM)

```python
scheme_a = qpam.QPAM()
```

```python
encoded_circuit_a = scheme_a.encode(original_data)
```

```python
encoded_circuit_a.draw('mpl')
```

```python
decoded_data = scheme_a.decode(encoded_circuit_a)
```

```python
utils.plot([original_data,decoded_data],title='QPAM',label=label)
```
