```python
import sys
sys.path.append("../src")
from schemes import qpam, sqpam, qsm
import utils
```

##### Example Data

```python
original_data = utils.simulate_data(num_samples=16)
utils.plot(original_data)
```

```python
utils.plot(original_data)
```

#### Quantum Representations

##### A) Quantum Probability Amplitude Modulation

```python
scheme_a = qpam.QPAM()
```

```python
encoded_circuit_a = scheme_a.encode(original_data)
```

```python
decoded_data    = scheme_a.decode(encoded_circuit_a)
```

```python
utils.plot([original_data,decoded_data])
```

```python
utils.tune(encoded_circuit_a,function=scheme_a.decode,step=2,max_value=4000,ref=original_data)
```
