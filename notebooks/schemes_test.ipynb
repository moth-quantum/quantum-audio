```python
import sys
sys.path.append("../src")
from schemes import qpam, sqpam, qsm
import utils
```

##### QPAM Example

```python
original_data = utils.simulate_data(num_samples=150)
scheme_1      = qpam.QPAM()
```

```python
encoded_circuit = scheme_1.encode(original_data)
decoded_data    = scheme_1.decode(encoded_circuit,shots=8000)
```

```python
utils.plot([original_data,decoded_data[:len(original_data)]])
```

```python
encoded_circuit.draw('mpl',style='clifford')
```

```python
utils.tune(encoded_circuit,function=scheme_1.decode,step=2,max_value=4000)
```
