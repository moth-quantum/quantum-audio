<h1 style="text-align: center;"> Example of a Digital Signal</h1>

```python
signal = test_signal()
```

<h1 style="text-align: center;"> Using Quantum Audio </h1>

```python
scheme = quantumaudio.SQPAM()
```

```python
circuit = scheme.encode(signal)
```

```python
output = scheme.decode(circuit)
plot([signal,output])
```

<h1 style="text-align: center;"> Revealing of Quantum Audio </h1>

```python
interactive.tune(circuit,function=scheme.decode,step=2,max_value=4000,ref=signal)
```
