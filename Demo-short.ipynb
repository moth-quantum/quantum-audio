### Signal Example

```python
signal = utils.simulate_data(8)
utils.plot(signal)
```

## Quantum Audio

```python
sqpam = quantumaudio.load('sqpam') 
```

```python
circuit = sqpam.encode(signal)
circuit.decompose().draw('mpl',style='clifford')
```

```python
utils.tune(circuit,function=sqpam.decode,step=2,max_value=4000,ref=signal)
```

#### Audio Example

```python
path = 'audio_samples/lead_guitar.mp3'
```

```python
qpam = quantumaudio.load('qpam')
```

#### <span style="color:tan">Original Digital Audio</span>

```python
y,sr = utils.get_chunks(file_path=path,sr=8000,mono=True,preview=True)
```

#### <span style="color:gold">Developing Quantum Audio from Noise</span>

```python
utils.tune_audio(obj=y,sr=sr,scheme=qpam,function=utils.process_chunks,limit=None,offset=0,max_value=4000)
```

#### Better Encoding Circuit

```python
qsm = quantumaudio.load('qsm')
```

```python
qsm.encode(signal).draw('mpl',style='clifford')
```

```python
utils.tune_audio(obj=y,sr=sr,scheme=qsm,function=utils.process_chunks,limit=None,offset=0,max_value=10000)
```
